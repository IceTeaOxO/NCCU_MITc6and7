# list的方法

## 基本（增加、減少）

在後方新增元素:append(想新增的元素)

```python
a=[1,2,3,4,5,6]
```

```python
a.append(10)
```

```python
a.append(11)
```

```python
a
```

插入元素：insert(要插入的位置,要插入的元素)

```python
a = [1,2,3,4,5]
```

```python
a.insert(1,"Hi")
```

```python
a
```

移除元素:remove(要移除的元素)

```python
a = [1,2,3,4,5,6]
```

```python
a.remove(4)
```

```python
a
```

移除某個位置的元素：pop(想移除的位置)

```python
a=[1,2,3,4,5]
```

```python
a.pop(3)
```

```python
a
```
