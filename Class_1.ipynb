input的型別為str，無法直接進行數字運算, 結果會是字串相加

```python
a = input()
b = input()
print(a+b)
type(a+b)
```

因此我們需要casting

```python
a = float(input())
b = float(input())
print(a+b)
print(a-b)
print(abs(a-b))
print(a*b)
print(a/b)
```
