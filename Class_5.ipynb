# 補充：實現階乘(5!=1x2x3x4x5=120)

```python
def factorial(n):
  if n<0:
    return "Wrong"
  elif n==0:
    return 1
  elif n==1:
    return 1
  else:
    return(n*factorial(n-1))
value=int(input())
print(value,"的階乘=",factorial(value))
print()
```
