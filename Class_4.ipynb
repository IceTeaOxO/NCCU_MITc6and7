巢狀迴圈

九九乘法表

```python
#巢狀迴圈:99乘法表
for i in range(1,10):
    for j in range(1,10):
       result=i*j
       print("%d*%d=%-3d"%(i,j,result),end="")
    print()#換行(1*1...換到2*1...)
```

直角三角形

```python
#巢狀迴圈:直角三角形
for i in range(1,10):
    for j in range(1,10):
        if j<=i:
            print("a",end="")
    print()
```

break：直接中斷迴圈

```python
for digit in range(1,11):
    if digit==6:
        print("斷")
        break
    print(digit,end=', ')
```

continue:先...後再繼續回圈

```python
scores=[11,22,33,44,0,55,60,66,77,1]
for score in scores:
    if score>=60:
        print("大於60")#score>60印這個(true)
        continue
    print(score)#score<=60印這個(false)
```

巢狀while寫成的巢狀99乘法表

```python
i=1
while i<=9:
    j=1
    while j<=9:
        result=i*j
        print("%d*%d=%d"%(i,j,result), end=" ")
        j+=1
    print()
    i+=1
```

驗證元素存在

```python
fruits={'apple':20,'orange':15,'guava':10}
```

```python
key=input("請輸入水果=")
value=int(input("請輸入價格="))
if key in fruits:
    print("%s有了"%key)
    if value!=fruits[key]:
        fruits[key]=value
else:
    print("%s原本沒有"%key)
    fruits[key]=value
    print("新的",fruits)
```

```python
fruits
```

求鍵、值

```python
#印出鍵、值->.items()
for name,price in fruits.items():
    print("\n水果: ",name)
    print("價格: ",price)
```

```python
#只要印出鍵->.keys()
for name in fruits.keys():
    print("\n水果: ",name)
```

```python
#只要印出值->.values()
for price in fruits.values():
    print("\n價格: ",price)
```

while在字典的應用

```python
survey={}
market=True
while market:
    country=input("\n請輸入國家: ")
    site=input("請輸入景點: ")
    survey[country]=site
    repeat=input("還有嗎?(y/n)")
    if repeat !='y':
        market=False
print("\n市調結果: ")
for cou,sit in survey.items():
    print(cou," 的 ",sit)
```

# 物件(補充)

Car:類別(class)

Toyota:Car之下的一個物件(object)

```python
class Car:
  def __init__(self,brand):
    self.brand=brand
  def Drive(self):
    print("Drive!")

class Motorcycle:
  def __init__(self,brand,whattype):
    self.brand=brand
    self.whattype=whattype
  def Ride(self):
    print("Ride!")
```

```python
Car1=Car("Toyata")
Motor1=Motorcycle("Gogoro","electric")
```

```python
Car1.Drive()
```

```python
Motor1.Ride()
```

```python
Car1.Ride()
```

```python
Car1.brand
```

```python
Motor1.whattype
```

```python
Motor1.type
```

```python
isinstance(Car1,Car)
```

```python
isinstance(Car1,Motorcycle)
```
